# Problem 3(a)

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time
%matplotlib inline

x = np.linspace(-1, 1, 100)
y1 = np.empty(len(x))
y2 = np.empty(len(x))
z1 = np.ones(len(x))
z2 = -np.ones(len(x))
y_classify = np.zeros(len(x)) - 0.05
for i in range(0, len(x)): 
    y1[i] = -0.6 * math.sin(np.pi/2 + 3*x[i]) - 0.35
    y2[i] = -0.6 * math.sin(np.pi/2 + 3*x[i]) + 0.25

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y_classify, linewidth=2.0, label='Linear Classifier')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel('X axis')
plt.ylabel('Y axis', rotation=0)
plt.title('The Two Curves', size=20)
plt.legend(loc='best')
plt.show()
```

# Problem 3(b)

## Grid search and visualize the transformed topology for several times

```python
plt_number = 0

try:
    for w11 in np.arange(-3, 3, 0.5):
        for w12 in np.arange(-3, 3, 0.5):
            for w21 in np.arange(-3, 3, 0.5):
                for w22 in np.arange(-3, 3, 0.5):
                    for b1 in np.arange(-1, 1, 0.5):
                        for b2 in np.arange(-1, 1, 0.5):
                            x_hat_1_plt = np.tanh(w11*x + w21*y1 + b1)
                            y_hat_1_plt = np.tanh(w12*x + w22*y1 + b2)
                            x_hat_2_plt = np.tanh(w11*x + w21*y2 + b1)
                            y_hat_2_plt = np.tanh(w12*x + w22*y2 + b2)
                            plt.plot(x_hat_1_plt, y_hat_1_plt)
                            plt.plot(x_hat_2_plt, y_hat_2_plt)
                            plt.xlabel('X axis')
                            plt.ylabel('Y axis', rotation=0)
                            plt.title('The Two Curves', size=20)
                            plt.show()
                            wait = input("PRESS ENTER TO CONTINUE.")
                            if plt_number==5:
                                raise NameError()
                            else:
                                plt_number += 1
                                plt.clf()
except NameError:
    pass
```

## Use the gradient descent method to minimize the loss function provided

```python
#data preprocessing
#concatenate 3 NumPy arrays by column-wise by specifying axis=1.
xvals1=np.concatenate((x[:, np.newaxis],y1[:, np.newaxis]), axis=1)
xvals2=np.concatenate((x[:, np.newaxis],y2[:, np.newaxis]), axis=1)
#concatenate 2 numpy arrays: row-wise
x_vals=np.concatenate((xvals1,xvals2), axis=0)
y_vals=np.concatenate((z1,z2), axis=0)
```

```python
import tensorflow.compat.v1 as tf
from pylab import mpl
tf.disable_v2_behavior()
# tf.disable_eager_execution() 
# np.random.seed(10)
# tf.set_random_seed(1)
sess=tf.Session()

#random suffle the training set
train_indices = np.random.choice(len(x_vals), size=len(x_vals), replace=False)
x_vals_train = x_vals[train_indices]
y_vals_train = y_vals[train_indices]

#define variable
x_data = tf.placeholder(shape=[None, 2], dtype=tf.float32)
y_target = tf.placeholder(shape=[None, 1], dtype=tf.float32)
W_x_hat = tf.Variable(tf.random_normal(shape=[2,1]))
b_x_hat = tf.Variable(tf.random_normal(shape=[1,1]))
W_y_hat = tf.Variable(tf.random_normal(shape=[2,1]))
b_y_hat = tf.Variable(tf.random_normal(shape=[1,1]))
#w=(a,b,c) a*x_hat + b*y_hat + c == t
a = tf.Variable(tf.random_normal(shape=[1,1]), constraint=lambda t: tf.clip_by_value(t, -1, 1))
b = tf.Variable(tf.random_normal(shape=[1,1]), constraint=lambda t: tf.clip_by_value(t, -1, 1))
c = tf.Variable(tf.random_normal(shape=[1,1]), constraint=lambda t: tf.clip_by_value(t, -1, 1))

#set the prediction as the linear function
tf_x = tf.matmul(x_data,W_x_hat) + b_x_hat
tf_y = tf.matmul(x_data,W_y_hat) + b_y_hat
tf_x_hat = tf.math.tanh(tf_x)
tf_y_hat = tf.math.tanh(tf_y)

#define the loss function
model_output = tf.multiply(a, tf_x_hat) + tf.multiply(b, tf_y_hat) + c
loss = tf.reduce_sum(-tf.multiply(model_output, y_target))
# loss = tf.reduce_mean(-tf.multiply(tf.math.log(model_output), y_target))

#定义模型输出。利用符号函数实现，结果大于0时，输出1，小于0时，输出-1。
prediction = tf.sign(model_output)
#定义准确率，利用的是一批样本的平均准确率。
accuracy = tf.reduce_mean(tf.cast(tf.equal(prediction, y_target),tf.float32))
#define Gradient Descent Optimizer
train_step=tf.train.GradientDescentOptimizer(0.05).minimize(loss)

#starting training and initialize all the variables
sess.run(tf.global_variables_initializer())
loss_vec = []
train_accuracy = []
test_accuracy = []
#set the number of iteration to 3000
for i in range(3000):
#Mini-batch Gradient Descent
#     rand_index = np.random.choice(len(x_vals_train), size=batch_size, replace=False)
#     rand_x = x_vals_train[rand_index]
#     rand_y = np.transpose([y_vals_train[rand_index]])
#Batch Gradient Descent
    rand_x = x_vals_train
    rand_y = y_vals_train[:, np.newaxis]
    _ , temp_loss = sess.run(fetches=[train_step, loss], feed_dict={x_data: rand_x, y_target: rand_y})
    loss_vec.append(temp_loss)
    if (i+1)%200==0:
        print('Step #' + str(i+1) + ' Loss = ' + str(temp_loss))
#         print('Step #' + str(i+1) + ' W = ' + str(sess.run(W)) + 'b = ' + str(sess.run(b)))

plt.plot(loss_vec, label='loss function')
plt.xlabel('The number of iterations')
plt.legend(loc='best')
plt.show()
```

## printing the results of 9 parameters
- W_x_hat corresponds to [w<sub>11</sub>, w<sub>12</sub>]
- W_y_hat corresponds to [w<sub>12</sub>, w<sub>22</sub>]
- b_x_hat corresponds to [b<sub>1</sub>]
- b_y_hat corresponds to [b<sub>2</sub>]
- w corresponds to [a, b, c]

```python
print(' W_x_hat = ' + str(sess.run(W_x_hat)) + '\n W_y_hat = ' + str(sess.run(W_y_hat)))
print(' b_x_hat = ' + str(sess.run(b_x_hat)) + '\n b_y_hat = ' + str(sess.run(b_y_hat)))
print(' a = ' + str(sess.run(a)) + ' b = ' + str(sess.run(b)) + ' c = ' + str(sess.run(c)))
```

```python
w1 = sess.run(W_x_hat)
w2 = sess.run(W_y_hat)
b1 = sess.run(b_x_hat)
b2 = sess.run(b_y_hat)
aa = np.squeeze(sess.run(a))
bb = np.squeeze(sess.run(b))
cc = np.squeeze(sess.run(c))
x_classifier = np.linspace(-1, 1, 100)
y_classifier = -aa*x_classifier/bb - cc/bb
x_hat_1 = np.tanh(w1[0][0]*x + w1[1][0]*y1 + b1)
y_hat_1 = np.tanh(w2[0][0]*x + w2[1][0]*y1 + b2)
x_hat_2 = np.tanh(w1[0][0]*x + w1[1][0]*y2 + b1)
y_hat_2 = np.tanh(w2[0][0]*x + w2[1][0]*y2 + b2)
plt.scatter(x_hat_1, y_hat_1, label='Class 1')
plt.scatter(x_hat_2, y_hat_2, label='Class 2')
plt.plot(x_classifier, y_classifier, color='red', label='Linear Classifier')
plt.xlabel('X axis')
plt.ylabel('Y axis', rotation=0)
plt.title('The Two Curves', size=20)
plt.legend(loc='best')
plt.show()

sess.close()
```

## In all cases examine sensitivity of the obtained parameter set with respect to the starting point

- The optimal parameter set is sensitive to starting points. 
- In my code, I used random standard normal to initialize all the parameters, which gives me a good classification result.
- However, if I initialize all the parameters with a constant value, such as 0 or 1, then the Gradient Descent Method would not converge and the gradient of the parameters would vanish during the training process.
- See the below for the detailed experiments when we initialize all the parameter to 1. 
- The result shows that w<sub>11</sub> = w<sub>12</sub>, w<sub>21</sub> = w<sub>22</sub>, b<sub>1</sub> = b<sub>2</sub>, which is not a good set of parameters.
- If we initialize all the parameter to 0, then the gradient of the parameters would vanish during the training process.
- Overall, it is better to initialize all the parameter set randomly normal as the starting point.
