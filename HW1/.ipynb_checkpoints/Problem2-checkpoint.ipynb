# Problem 2(a)
- plot the two spirals

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

t = np.arange(1, 101)
r1= 60 + 0.20*t
r2 = 40 + 0.40*t
Phi1 = -0.06*t + 3
Phi2 = -0.08*t + 3

x1 = np.empty(len(Phi1))
y1 = np.empty(len(Phi1))
z1 = np.ones(len(Phi1))
x2 = np.empty(len(Phi2))
y2 = np.empty(len(Phi2))
z2 = -np.ones(len(Phi1))
for i in range(0, len(Phi1)): 
    x1[i] = r1[i] * math.cos(Phi1[i])
    y1[i] = r1[i] * math.sin(Phi1[i])
    x2[i] = r2[i] * math.cos(Phi2[i])
    y2[i] = r2[i] * math.sin(Phi2[i])
    
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.xlim(-100,100)
plt.ylim(-100,100)
plt.xlabel('X axis')
plt.ylabel('Y axis', rotation=0)
plt.title('The Two Spirals', size=20)
plt.show()
```

# Problem 2(b)
- for classification, transform the 2-dim space to a 3-dim space and find a hyperplane to classify, use hinge loss to achieve this

```python
fig = plt.figure()
ax1 = plt.axes(projection='3d')
ax1.plot3D(x1, y1, Phi1, label='1st Spiral', color='red', linewidth=2.5)     #绘制空间曲线
ax1.plot3D(x2, y2, Phi2, label='2nd Spiral', color='blue', linewidth=2.5)    #绘制空间曲线
ax1.legend(loc='best')
plt.xlim(-100,100)
plt.ylim(-100,100)
plt.show()
```

```python
#data preprocessing
#concatenate 3 NumPy arrays by column-wise by specifying axis=1.
xvals1=np.concatenate((x1[:, np.newaxis],y1[:, np.newaxis],Phi1[:, np.newaxis]), axis=1)
xvals2=np.concatenate((x2[:, np.newaxis],y2[:, np.newaxis],Phi2[:, np.newaxis]), axis=1)
#concatenate 2 numpy arrays: row-wise
x_vals=np.concatenate((xvals1,xvals2), axis=0)
y_vals=np.concatenate((z1,z2), axis=0)
```

```python
import tensorflow.compat.v1 as tf
from sklearn import datasets
from pylab import mpl
tf.disable_v2_behavior()
# tf.disable_eager_execution() 
np.random.seed(1)
tf.set_random_seed(1)
sess=tf.Session()

#random split the train and the test set
train_indices = np.random.choice(len(x_vals), size=round(len(x_vals)*0.75), replace=False)
test_indices = np.array(list(set(range(len(x_vals))) - set(train_indices)))
x_vals_train = x_vals[train_indices]
x_vals_test = x_vals[test_indices]
y_vals_train = y_vals[train_indices]
y_vals_test = y_vals[test_indices]

#set mini batch size
batch_size = 20
x_data = tf.placeholder(shape=[None, 3], dtype=tf.float32)
y_target = tf.placeholder(shape=[None, 1], dtype=tf.float32)
W = tf.Variable(tf.random_normal(shape=[3,1]))
b = tf.Variable(tf.random_normal(shape=[1,1]))

#set the prediction as the linear function
model_output = tf.matmul(x_data,W) + b
l2_norm = tf.reduce_sum(tf.square(W))
#set L2 Norm coefficient to avoid overfitting
alpha = tf.constant([0.01])
#define hinge loss function with L2 regularization
classification_term = tf.reduce_mean(tf.maximum(0., 1.-tf.multiply(model_output, y_target)))
loss = classification_term + alpha*l2_norm
#定义模型输出。利用符号函数实现，结果大于0时，输出1，小于0时，输出-1。
prediction = tf.sign(model_output)
#定义准确率，利用的是一批样本的平均准确率。
accuracy = tf.reduce_mean(tf.cast(tf.equal(prediction, y_target),tf.float32))
#define Gradient Descent Optimizer
train_step=tf.train.GradientDescentOptimizer(0.008).minimize(loss)

#starting training and initialize all the variables
sess.run(tf.global_variables_initializer())
loss_vec = []
#set the number of iteration to 50
for i in range(50):
#Mini-batch Gradient Descent
#     rand_index = np.random.choice(len(x_vals_train), size=batch_size, replace=False)
#     rand_x = x_vals_train[rand_index]
#     rand_y = np.transpose([y_vals_train[rand_index]])
#Batch Gradient Descent
    rand_x = x_vals_train
    rand_y = y_vals_train[:, np.newaxis]
    _ , temp_loss = sess.run(fetches=[train_step, loss], feed_dict={x_data: rand_x, y_target: rand_y})
    loss_vec.append(temp_loss)
    if (i+1)%10==0:
        print('Step #' + str(i+1) + '\n W = ' + str(sess.run(W)) + '\n b = ' + str(sess.run(b)))
        print(' Loss = ' + str(temp_loss))

plt.plot(loss_vec, label='loss function')
plt.xlabel('The number of iterations')
plt.legend(loc='best')
plt.show()
```

```python
train_acc_temp = sess.run(accuracy, feed_dict={x_data: x_vals_train, y_target: np.transpose([y_vals_train])})
test_acc_temp = sess.run(accuracy, feed_dict={x_data: x_vals_test, y_target: np.transpose([y_vals_test])})
print('Training accuracy is', train_acc_temp)
print('Testing accuracy is', test_acc_temp)
```

```python
ww = sess.run(W)
bb = np.squeeze(sess.run(b))
xx = np.arange(-100, 100, 0.5)
yy = np.arange(-100, 100, 0.5)
x_classifier, y_classifier = np.meshgrid(xx, yy)
z_classifier = -(ww[0][0]*x_classifier + ww[1][0]*y_classifier + bb)/ww[2][0]

#define a new 3-D figure
fig2 = plt.figure()
ax3 = plt.axes(projection='3d')
ax3.plot_surface(x_classifier, y_classifier, z_classifier, cmap='rainbow', zorder=1)
ax3.plot3D(x1, y1, Phi1, label='1st Spiral', color='red', linewidth=2.0, zorder=2)
ax3.plot3D(x2, y2, Phi2, label='2nd Spiral', color='blue', linewidth=2.0, zorder=3)
ax3.legend(loc='best')
plt.xlim(-100,100)
plt.ylim(-100,100)
plt.show()

# sess.close()
```
